Restricting aircraft category to airplanes, but including many rows that had a missing value for aircraft category if it had the same make and model as another row specified as an Airplane.

```python
airplane_df = df[df['Aircraft.Category']=='Airplane']
```

```python
airplane_make_list = [make for make in airplane_df['Make']]
```

```python
airplane_model_list = [model for model in airplane_df['Model']]
```

```python
def replace_airplane(row):
    if pd.isnull(row['Aircraft.Category']) and row['Make'] in airplane_make_list and row['Model'] in airplane_model_list:
        return 'Airplane'
    else:
        return row['Aircraft.Category']


df['Aircraft.Category'] = df.apply(replace_airplane, axis=1)
```

Drop NaN values in 'Model' column.

Impute correct Number of Engines based on similar data with same make and model.

```python
clean_df['Number.of.Engines'].isna().sum()
```

```python
clean_df['Number.of.Engines'].value_counts()
```

```python
clean_df.shape
```

```python
clean_df = clean_df.reset_index()
```

```python
num_eng_dict = {}
for i in range(clean_df.shape[0]):
    if (clean_df.loc[i,'Make']+' '+clean_df.loc[i,'Model']) not in num_eng_dict.keys():
        num_eng_dict[(clean_df.loc[i,'Make']+' '+clean_df.loc[i,'Model'])] = clean_df.loc[i,'Number.of.Engines']
```

```python
num_eng_dict
```

```python
def replace_num_eng(row):
    if pd.isnull(row['Number.of.Engines']) and (row['Make']+' '+row['Model']) in num_eng_dict.keys():
        return num_eng_dict[(row['Make']+' '+row['Model'])]
    else:
        return row['Number.of.Engines']


clean_df['Number.of.Engines'] = clean_df.apply(replace_num_eng, axis=1)
```

```python
clean_df['Number.of.Engines'].value_counts()
```

```python
clean_df['Number.of.Engines'].isna().sum()
```

Dropping remaining missing Number of Engine rows.

```python
clean_df = clean_df.dropna(subset=['Number.of.Engines']).copy()
```

```python
clean_df['Number.of.Engines'].isna().sum()
```

Following a similar process for Engine Type:

```python
clean_df['Engine.Type'].isna().sum()
```

```python
clean_df = clean_df.reset_index()
```

```python
eng_type_dict = {}
for i in range(clean_df.shape[0]):
    if (clean_df.loc[i,'Make']+' '+clean_df.loc[i,'Model']) not in eng_type_dict.keys():
        eng_type_dict[(clean_df.loc[i,'Make']+' '+clean_df.loc[i,'Model'])] = clean_df.loc[i,'Engine.Type']
```

```python
eng_type_dict
```

```python
def replace_eng_type(row):
    if pd.isnull(row['Engine.Type']) and (row['Make']+' '+row['Model']) in eng_type_dict.keys():
        return eng_type_dict[(row['Make']+' '+row['Model'])]
    else:
        return row['Engine.Type']


clean_df['Engine.Type'] = clean_df.apply(replace_eng_type, axis=1)
```

```python
clean_df['Engine.Type'].isna().sum()
```

```python
clean_df['Engine.Type'].value_counts()
```

```python
#Dropping remaining missing engine types
clean_df = clean_df.dropna(subset=['Engine.Type']).copy()
```

```python
clean_df['Engine.Type'].isna().sum()
```
